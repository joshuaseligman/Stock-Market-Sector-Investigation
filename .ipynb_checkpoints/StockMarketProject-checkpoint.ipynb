In this project, I used data analysis tools in Python to collect and analyze stock market data from 2020. I was particularly interested in this project because of my understanding of the relationships between different industries and I wanted to see if there are any trends within the stock prices of companies in various industries.

```python
import yfinance as yf
import numpy as np
import pandas
import matplotlib
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import datetime
```

In the cell below, I collected the stock market symbols of the companies within the S&P 500. To complete this task, I web scraped the Wikipedia page with the list of the S&P 500 companies using the `BeautifulSoup` module. The `contents` variable is a list of the rows within the table on the Wikipedia site.

```python
URL = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
site = requests.get(URL)
soup = BeautifulSoup(site.content, 'html.parser')
table = soup.find(id='constituents')
contents = table.find_all('tr')[1:]
print(contents[0])
```

This cell collects the stock data for each company within the S&P 500. I began by declaring a dictionary with keys for each sector that point to empty lists. I then iterated through each company, collected the open and close prices for each day, and then created a new DataFrame object for each company that reflects the percent change in the price for each day. I then added each company's stock data to the list for the industry to which they belonged to.

```python
#Create the dictionary for each sector
sector_stock_data = {'Communication Services': [], 'Consumer Discretionary': [], 'Consumer Staples': [], \
                     'Energy': [], 'Financials': [], 'Health Care': [], 'Industrials': [], \
                     'Information Technology': [], 'Materials': [], 'Real Estate': [], 'Utilities': []}

#Define start and end dates
start_date = '2020-1-1'
end_date = '2021-1-1'

#Iterate through each row in the table
for row in contents:
    #Get a list of the columns and get the name and sector
    cols = row.find_all('td')
    name = cols[0].find('a').text.replace('.', '-')
    sector = cols[3].text.replace('\n', '')
    
    #Create the ticker object and get the data
    ticker = yf.Ticker(name)
    stock_data = ticker.history(period='1d', start=start_date, end=end_date)
    
    #Manipulate the data
    open_data = stock_data['Open']
    close_data = stock_data['Close']
    percent_change = (close_data - open_data) / open_data * 100
    volume = stock_data['Volume']
    new_dataframe = pandas.DataFrame(data={'Percent Change': percent_change, 'Volume': volume})
    
    #Add the data to the appropriate sector
    sector_stock_data[sector].append([name, new_dataframe])
```

This cell tests the results of the previous cell by graphing the percent change in Amazon's stock throughout the year.

```python
random_data = sector_stock_data['Consumer Discretionary'][1]
print(random_data[0])
plt.plot(random_data[1]['Percent Change'])

plt.xlabel('Date', fontsize = 15)
plt.ylabel('Percent Change', fontsize = 15)

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 50
plt.show()
```

The cell below contains a majority of the calculations that I made throughout the project. The main part of my investigation is related to the performance of each sector. Since I had the data for each company, I had to now convert this data into something that reflects the overall performance of each sector. I computed the total volume of each company's stock for each day of trading. For each sector, I began by calculating the sum of the volumes of each company's stock within the sector for every day in the data. With this sum, I was then able to compute a weighted average of the percent changes of the companies in relation to their volumes. I chose to use the company's daily stock volume to compare them with the other companies in the sector because a company with more volume carries a larger weight in analyzing the performance of the sector on any given day. The computed average for the percent change in a sector represents the approximate overall performance of the sector in a given day.

```python
#Get the index for the days a create a new DataFrame with the index for each day
days_index = sector_stock_data['Materials'][0][1].index
sector_weighted_percent_change = pandas.DataFrame(index = days_index)

for sector in sector_stock_data:
    #Iterate sum up the volume of each company for the day
    volume_sums = pandas.Series(index = days_index, dtype = np.float).fillna(0)
    for company in sector_stock_data[sector]:
        #Fix dataframe if missing data
        if company[1].shape[0] < days_index.size:
            new_df = pandas.DataFrame(index = days_index, data = company[1])
            new_df = new_df.fillna(0)
            company[1] = new_df
        #Add to volume
        volume_sums += company[1]['Volume']
    
    #Create the series for the final data
    sector_weighted_percent_change[sector] = pandas.Series(index = days_index, dtype = np.float).fillna(0)
    
    #Compute the sector's weighted percent change for each day
    for company in sector_stock_data[sector]:
        sector_weighted_percent_change[sector] += company[1]['Percent Change'] * company[1]['Volume'] / volume_sums
#Output the sector percent changes
print(sector_weighted_percent_change.head())
```

I analyzed the performance of the sectors in 2 different conditions. The first of which is displayed in the cell below where I compared the number of times each sector saw the largest rise and fall compared to the other sectors each day. I was shocked by how noticeable the patterns ended up being for this analysis. Sectors such as Consumer Staples, Financials, and Health Care were all among the sectors that were infrequent in performing the best or worst in the market. My prediction for the reasoning behind this result is that these sectors are necessities in the economy and are essential to consumers and companies. In other words, investors can safely invest in these companies because they are going to perform at a somewhat consistent rate over time. On the other hand, sectors like Energy, Consumer Discretionary, and Information Technology were consistent in either performing well or poorly. These companies are not always the most reliable because the variables that drive the performance of these companies (i.e., oil prices, consumer tastes, and technology) are constantly changing and, therefore, companies have to be able to adapt to the changes in the marketplace to consistently perform well. Additionally, since this data is from 2020 during the COVID-19 pandemic, consumers lost a lot of money from the recession and had less disposable income to spend on luxury goods, which caused many companies in the Consumer Discretionary sector to take many losses.

```python
stocks_tops_bottoms = pandas.DataFrame(index = ['Consumer Discretionary', 'Consumer Staples', \
                                                'Industrials', 'Materials', 'Communication Services', \
                                                'Financials', 'Health Care', 'Energy', \
                                                'Information Technology', 'Utilities', 'Real Estate'], \
                                       columns = ['Max', 'Min']).fillna(0)

for day, day_info in sector_weighted_percent_change.iterrows():
    min_value = day_info.min()
    max_value = day_info.max()
    for company in stocks_tops_bottoms.index:
        if day_info[company] == min_value:
            stocks_tops_bottoms['Min'][company] += 1
        if day_info[company] == max_value:
            stocks_tops_bottoms['Max'][company] += 1
    
print()
print(stocks_tops_bottoms)

label_locations = np.arange(len(stocks_tops_bottoms.index))
bar_width = 0.35
fig, ax = plt.subplots()
max_bars = ax.bar(label_locations - bar_width / 2, stocks_tops_bottoms['Max'], width = bar_width, label = 'Max')
min_bars = ax.bar(label_locations + bar_width / 2, stocks_tops_bottoms['Min'], width = bar_width, label = 'Min')

ax.set_xticks(label_locations)
ax.set_xticklabels(stocks_tops_bottoms.index, fontsize = 10)
ax.legend(prop = {'size': 25})

plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Count', fontsize = 15)

plt.rcParams['figure.figsize'] = [25, 8]
plt.rcParams['figure.dpi'] = 150
plt.show()
```

The second half of my investigation involved the comparison of sectors that relate to each other. More specifically, I compared the performances of the Consumer Discretionary, Consumer Staples, Industrials, and Materials sectors. With this analysis, I wanted to find out if the performance of consumer-based sectors has an association with the performance of sectors that act as suppliers for these consumer-based sectors. Unfortunately, no association was able to be found as all of the sectors, except for Consumer Staples, are constantly going back and forth in being the best and worst-performing sectors. For instance, Consumer Discretionary tops these sectors around January 12 but is right in the middle around January 23 and vice versa for the Materials sector for the same dates. As previously mentioned, Consumer Staples is the only sector that did not follow this trend as it was consistent in being an average performer throughout the entire month of January. The findings from the previous investigation may be related to this result. However, the fluctuations of the stock market do not give me enough evidence to believe that these results are entirely related.

```python
desired_sectors = ['Consumer Discretionary', 'Consumer Staples', 'Materials', 'Industrials']
for sector in sector_weighted_percent_change:
    if desired_sectors.count(sector) != 0:
        plt.plot(sector_weighted_percent_change[sector], label=sector)
        plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left', prop = {'size': 10})
plt.xlim([datetime.date(2020, 1, 1), datetime.date(2020, 1, 31)])
plt.ylim([-3, 3])

plt.xlabel('Date', fontsize = 15)
plt.ylabel('Percent Change', fontsize = 15)

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100
plt.show()
```
